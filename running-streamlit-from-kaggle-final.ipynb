```python
%%writefile app.py

import streamlit as st
import easyocr
import matplotlib.pyplot as plt
from huggingface_hub import login
from PIL import Image
import pytesseract
from langchain import HuggingFacePipeline
from transformers import AutoTokenizer
import transformers
from time import time
import torch
import transformers
import pandas as pd
import nltk
import keras_ocr
import numpy as np
from bokeh.models.widgets import Button
from bokeh.models import CustomJS
from streamlit_bokeh_events import streamlit_bokeh_events
from langchain import PromptTemplate,  LLMChain
from langchain.prompts.prompt import PromptTemplate
from langchain.prompts.few_shot import FewShotPromptTemplate
import streamlit as st
from getpass import getpass
import os
import nest_asyncio
from langchain.llms import HuggingFaceHub
from langchain.document_loaders import WebBaseLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceInferenceAPIEmbeddings
from langchain_community.document_loaders import TextLoader
from langchain.vectorstores import Chroma
from langchain.chains import RetrievalQA, LLMChain
from huggingface_hub import login


st.set_page_config(page_title="Notezy", layout='wide',
                   page_icon="📖", )

generated_notes = ""



rad = st.sidebar.selectbox('Navigation Bar', ['Home', 'Text extraction', 'Final pipline', 'QandA'])


  

if rad == 'Home':
#     set_background(r"C:\Users\abhin\Downloads\background.jpeg")
    st.title('Notezy')
    # st.image(r'D:\abhin\Comding\ML\AspectWise\Streamlit\headphones.jpg')
    st.header('- A comprehensive note-taking companion')
    st.subheader('About')
    st.text('''Notezy is a comprehensive note making solution trying to revolutionize
the education industry by automating the process of note-making for students. This
helps students focus more on the lecture than simultaneously be bombarded with note-making.
Students can also have personalized notes based on the examples notes provided 
to the model. 

The working of Notezy includes a basic pipeline with an OCR model for text extraction 
combined with LLM on top of it to generate notes on the given topic. The particular implementation
uses EasyOCR and Mistral-7b as the backend models.
            ''')

if rad == 'Text extraction':
    text_models = st.sidebar.radio('Extraction models', ['Tesseract', 'Easy OCR', 'Keras OCR', 'Comparison'])
    
    if text_models == 'Easy OCR':
        st.title("Text extraction Playground")
        st.header('Easy OCR')
        reader = easyocr.Reader(['en'], gpu=True)

        uploaded_file = st.file_uploader("Upload an image", type=["jpg", "jpeg", "png"])

        if uploaded_file is not None:
            st.image(uploaded_file, caption='Uploaded Image')

        button = st.button('Recognize text')
        if button:
            image = Image.open(uploaded_file)
            image = np.array(image)
            results = reader.readtext(image)
            st.write('Text Extracted : ')

            extracted = ''
            for _, text, prob in results:
                extracted = extracted + ' ' + text
                print('Text : ', text)

            st.success(extracted)  
        
    if text_models == 'Tesseract':
        custom_config = r'--oem 3 --psm 6'
        st.title("Text extraction Playground")
        st.header('Tesseract OCR')

        uploaded_file = st.file_uploader("Upload an image", type=["jpg", "jpeg", "png"])

        if uploaded_file is not None:
            st.image(uploaded_file, caption='Uploaded Image')

        button = st.button('Recognize text')
        if button:
            image = Image.open(uploaded_file)
            extracted = pytesseract.image_to_string(image, config=custom_config)
            st.success(extracted)
      
    if text_models == 'Keras OCR':
        st.title("Text extraction Playground")
        st.header('Keras OCR')
        uploaded_file = st.file_uploader("Upload an image", type=["jpg", "jpeg", "png"])

        pipeline = keras_ocr.pipeline.Pipeline()
        if uploaded_file is not None:
            st.image(uploaded_file, caption='Uploaded Image')

#         images = [keras_ocr.tools.read(img) for img in [uploaded_file]]
#         prediction_groups = pipeline.recognize(images)
        
        button = st.button('Recognize text')
        if button:
            image = Image.open(uploaded_file)
            image_array = np.array(image)
            images = [image_array]
            prediction_groups = pipeline.recognize(images)
            predicted_image = prediction_groups[0]
            extracted = ''
            for text, box in predicted_image:
                extracted = extracted + ' ' + text
            st.success(extracted)
            
    if text_models == 'Comparison':
        st.title("Text extraction Playground")
        st.header('Comparitive analysis of models')
        
        uploaded_file = st.file_uploader("Upload an image", type=["jpg", "jpeg", "png"])
        og_text = st.text_area('Enter Actual text for comparison')
        
        pipeline = keras_ocr.pipeline.Pipeline()
        reader = easyocr.Reader(['en'], gpu=False)
        custom_config = r'--oem 3 --psm 6'
        
        if uploaded_file is not None:
            st.image(uploaded_file, caption='Uploaded Image')
        
        button = st.button('Recognize text')
        if button:
            image = Image.open(uploaded_file)
            image_array = np.array(image)
            
            #Easy OCR
            start = time()
            easyocr = reader.readtext(image)
            end = time()  
            extracted_easyocr = ''
            for _, text, prob in easyocr:
                extracted_easyocr = extracted_easyocr + ' ' + text
            time_easyocr = round(end - start, 2)
            
            lev_distance_easyocr = nltk.edit_distance(og_text, extracted_easyocr)
            with st.expander('Easy OCR - Text Extracted'):
                st.write(extracted_easyocr)
                st.write(f'Levenstein Distance : {lev_distance_easyocr}')
                st.write(f'Time taken : {time_easyocr}')
            
            #Tesseract
            start = time()
            extracted_tes = pytesseract.image_to_string(image, config=custom_config)
            end = time()
            time_tes = round(end - start, 2)
            
            lev_distance_tes = nltk.edit_distance(og_text, extracted_tes)
            with st.expander('Tesseract OCR - Text Extracted'):
                st.write(extracted_tes)
                st.write(f'Levenstein Distance : {lev_distance_tes}')
                st.write(f'Time taken : {time_tes}')
                
            #Keras OCR
            start = time()
            images = [image_array]
            prediction_groups = pipeline.recognize(images)
            predicted_image = prediction_groups[0]
            extracted_ker = ''
            for text, box in predicted_image:
                extracted_ker = extracted_ker + ' ' + text
            end = time()
            time_ker = round(end - start, 2)
            
            lev_distance_ker = nltk.edit_distance(og_text, extracted_ker)
            with st.expander('Keras OCR - Text Extracted'):
                st.write(extracted_ker)   
                st.write(f'Levenstein Distance : {lev_distance_ker}')
                st.write(f'Time taken : {time_ker}')
        

            data = {'OCR Method': ['Easy OCR', 'Tesseract OCR', 'Keras OCR'],
                    'Levenshtein Distance': [lev_distance_easyocr, lev_distance_tes, lev_distance_ker],
                    'Time taken' : [time_easyocr, time_tes, time_ker]}
            df = pd.DataFrame(data)
            
            fig, axs = plt.subplots(1, 2, figsize=(15, 5))  # Create 1 row, 2 columns of subplots

            # Plot Levenshtein Distance
            axs[0].bar(df['OCR Method'], df['Levenshtein Distance'])
            axs[0].set_title('Levenshtein Distance for OCR Methods (lower is better)')
            axs[0].set_xlabel('OCR Method')
            axs[0].set_ylabel('Levenshtein Distance')

            # Plot Time Taken
            axs[1].bar(df['OCR Method'], df['Time taken'])
            axs[1].set_title('Time Taken for OCR Methods')
            axs[1].set_xlabel('OCR Method')
            axs[1].set_ylabel('Time taken')

            # Display the charts in Streamlit
            st.pyplot(fig)
            
            
if rad == 'Final pipline':
    st.title("Final Pipeline Playground")

    uploaded_file = st.file_uploader("Upload an image", type=["jpg", "jpeg", "png"])

    if uploaded_file is not None:
        st.image(uploaded_file, caption='Uploaded Image')

    extracted = ''
    button1 = st.button('Recognize text and generate notes')
    if button1:
        reader = easyocr.Reader(['en'], gpu=False)
        image = Image.open(uploaded_file)
        image = np.array(image)
#         st.write(uploaded_file)
        results = reader.readtext(image)
        st.write('Text Extracted : ')

        for _, text, prob in results:
            extracted = extracted + ' ' + text
            print('Text : ', text)

        st.success(extracted)

    #     button2 = st.button('Generate notes')
    #     if button2:
        token='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'
        login(token)

        model = "meta-llama/Llama-2-7b-chat-hf"
        tokenizer = AutoTokenizer.from_pretrained(model, token=token)

        pipeline = transformers.pipeline(
            "text-generation", #task
            model=model,
            tokenizer=tokenizer,
            torch_dtype=torch.bfloat16,
            trust_remote_code=True,
            device_map="auto",
            max_length=1500,
            do_sample=True,
            top_k=10,
            token=token,
            num_return_sequences=1,
            eos_token_id=tokenizer.eos_token_id
        )

        llm = HuggingFacePipeline(pipeline = pipeline, model_kwargs = {'temperature':0.2})

        template1 = """"You have to generate structured notes on the given text, Text :  {text}"""

        examples  = [
            {
                "text" : 'context : Machine learning',
                "notes" : '''
        'Machine Learning Basics'
        Notes:
        - Machine Learning (ML) is a subset of Artificial Intelligence (AI), focusing on algorithms that enable computers to learn from and make predictions or decisions based on data.
        - Various algorithms are utilized within machine learning, each serving different purposes and suited for different types of data.
        - Many machine learning algorithms are based on statistical principles, leveraging mathematical models to analyze and interpret data.
        - Examples of ML algorithms include logistic regression, linear regression, Support Vector Machines (SVM), k-nearest neighbors (KNN), and Naive Bayes, among others.
        - Data pre-processing and feature selection are crucial steps in ML, as they involve preparing and cleaning the data to ensure optimal performance of the algorithms.
        - ML algorithms are particularly effective for handling numerical data, though they can also be applied to categorical and textual data with appropriate preprocessing.
        - Different algorithms are chosen based on the specific use case and the nature of the data being analyzed, as each algorithm may excel in certain scenarios while underperforming in others.

                '''
            },
            {
                'text' : 'Context : Cloud Computing',
                'notes': '''
        'Cloud Computing Overview'
        Notes:
        -  What is Cloud Computing?:
        Cloud computing refers to the delivery of computing services, including servers, storage, databases, networking, software, and analytics, over the internet.

        - Cloud computing is utilized across different service models, including:
        Platform as a Service (PaaS): Provides a platform for developers to build, deploy, and manage applications without worrying about infrastructure.
        Infrastructure as a Service (IaaS): Offers virtualized computing resources, such as virtual machines and storage, on-demand.
        Software as a Service (SaaS): Delivers software applications over the internet, accessible via web browsers.
        - Big Data Analytics, Internet of Things (IoT), Artificial Intelligence (AI), and more.

        - Importance and Adoption:
        Cloud computing has gained significant importance in recent years due to its flexibility, scalability, and cost-effectiveness.
        Many businesses are migrating their operations to the cloud to leverage its benefits and improve efficiency.

        - Advantages: Scalability , Flexibility , Cost Savings , Accessibility , Reliability etc
        - Disadvantages: Security , Dependency , Compliance Challenge etc'''

            }

            ]
        few_shot_prompt_template = """"{text}"""
        example_prompt = PromptTemplate(
            input_variables=["text", "notes"], template= "Input_text: {text}\n{notes}"
        )
        prompt = FewShotPromptTemplate(
        examples=examples,
        example_prompt=example_prompt,
        suffix=few_shot_prompt_template,
        input_variables=["text"],
        )
        
        llm_chain = LLMChain(prompt=prompt, llm=llm)
        text = f'Context : {extracted}'
        st.write(text)
        
        
        start_time = time.time()
        output = llm_chain.run(text.lower())
#         st.write(output)
        end_time = time.time()
        ttft = end_time - start_time
        total_tokens = len(output.split())
        tpot = (end_time - start_time - ttft) / total_tokens

        latency = ttft + (tpot * total_tokens)
        print(f"Time taken per token: {tpot:.4f} seconds")
        print(f"Overall latency: {latency:.4f} seconds")
#         st.write(output)
        notes = output[output.index(text.lower()):].replace(text.lower(), '').strip('\n')
        generated_notes = notes
        st.write(notes)
        
        
        st.subheader("Notes guide")
        user_input = st.text_input("Enter your query:")

        if st.button("Submit"):
            st.write("Loading...")

            login('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')
            HF_TOKEN = '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'
            nest_asyncio.apply()
            file_name = "my_text_file.txt"

            # Write the text to the file
            with open(file_name, "w") as file:
                file.write(generated_text)
        #     WEBSITE_URL = "https://www.ibm.com/topics/machine-learning#:~:text=Machine%20learning%20(ML)%20is%20a,learn%2C%20gradually%20improving%20its%20accuracy."
            loader = TextLoader("/kaggle/input/mltext/my_text_file.txt")
            docs = loader.load()
        #     loader = WebBaseLoader(WEBSITE_URL)
        #     loader.requests_per_second = 1
        #     docs = loader.aload()

            text_splitter = RecursiveCharacterTextSplitter(chunk_size=256, chunk_overlap=20)
            chunks = text_splitter.split_documents(docs)

            embeddings = HuggingFaceInferenceAPIEmbeddings(api_key=HF_TOKEN, model_name="BAAI/bge-base-en-v1.5")
            vectorstore = Chroma.from_documents(chunks, embeddings)

            search = vectorstore.similarity_search(user_input)
            retriever = vectorstore.as_retriever(search_type="mmr", search_kwargs={'k': 4})
            relevant_docs = retriever.get_relevant_documents(user_input)


            prompt = f"""
                    <|system|>>
                    You are an AI QnA Assistant that follows instructions extremely well.
                    Be truthful and give direct answers. Please tell 'I don't know' if user query is not in context
                    </s>
                    <|user|>
                    {user_input}
                    </s>
                    <|assistant|>
                    """
            from langchain.llms import HuggingFaceHub
            from langchain.chains import RetrievalQA

            model = HuggingFaceHub(repo_id="HuggingFaceH4/zephyr-7b-alpha", huggingfacehub_api_token=HF_TOKEN,
                                   model_kwargs={"temperature":0.5,
                                                 "max_new_tokens":512,
                                                 "max_length":64
                                                })

            qa = RetrievalQA.from_chain_type(llm=model,retriever=retriever,chain_type="stuff")
            st_time = time.time()
            response = qa(prompt)
            nd_time = time.time()
            print("QnA retrieval time:", nd-time - st_time)
        #     prompt = f"You are an AI assistant that follows instructions extremely well. Please be truthful and give direct answers\n\n{user_input}\n"
        #     response = RetrievalQA.from_chain_type(llm=llm, chain_type="refine", retriever=retriever).run(prompt)
            output = response['result']
            output = output[output.index('Helpful Answer'):]
            st.write(output)
```
