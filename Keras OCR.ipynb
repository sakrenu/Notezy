```python
import keras_ocr
```

```python
pipeline = keras_ocr.pipeline.Pipeline()
```

```python
images = [
    keras_ocr.tools.read(img) for img in ['/content/drive/MyDrive/SC/Project/Data/Handwriting/Thermodynamics-handwritten.jpeg',
                                          '/content/drive/MyDrive/SC/Project/Data/Handwriting/ML-handwritten.jpeg',
                                          '/content/drive/MyDrive/SC/Project/Data/Handwriting/ML-text.png'
                                          ]
]
```

```python
prediction_groups = pipeline.recognize(images)
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(nrows=len(images), figsize=(10, 20))
for ax, image, predictions in zip(axs, images, prediction_groups):
    keras_ocr.tools.drawAnnotations(image=image,
                                    predictions=predictions,
                                    ax=ax)
```
