### Sample 1

```python
img = cv2.imread(r"/kaggle/input/image-forocr/WhatsApp Image 2024-01-20 at 3.44.45 PM.jpeg")

if img is None:
    print("Image not loaded properly. Check the path.")
else:
    plt.imshow(img)
```

```python
results = reader.readtext('/kaggle/input/image-forocr/WhatsApp Image 2024-01-20 at 3.44.45 PM.jpeg')

sum_prob = 0

for (bbox, text, prob) in results:
    print(f'Text: {text}, Probability: {prob}')
    sum_prob += prob
    
sum_prob / len(results)
```
